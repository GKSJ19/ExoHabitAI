```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('exoplanetai_datacleaning3_handlenullWithoutMedian.csv')
```

```python
# Low count physical essentials
low_null_cols = ['pl_orbper', 'pl_orbsmax', 'pl_masse', 'pl_msinie', 
                 'pl_cmasse', 'pl_bmasse', 'st_mass', 'pl_rvamp']

for col in low_null_cols:
    df[col] = df[col].fillna(df[col].median())
```

```python
# Physical characteristics
phys_cols = ['pl_rade', 'pl_dens', 'pl_insol', 'pl_eqt', 'st_rad', 
             'st_lum', 'st_logg', 'st_age', 'st_dens']

for col in phys_cols:
    df[col] = df[col].fillna(df[col].median())
```

```python
# Geometric and observational timing
geom_cols = ['pl_orbincl', 'pl_tranmid', 'pl_trandep', 'pl_trandur', 
             'pl_ratdor', 'pl_ratror', 'pl_orbtper', 'pl_orblper']

for col in geom_cols:
    df[col] = df[col].fillna(df[col].median())
```

```python
# Spatial coordinates
dist_cols = ['sy_dist', 'sy_plx']

for col in dist_cols:
    df[col] = df[col].fillna(df[col].median())
```

```python
df.isnull().sum().sum()
```

```python
df.to_csv('exoplanetai_datacleaned4_final_withMedian.csv', index=False)
```
