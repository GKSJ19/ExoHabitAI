```python
df_col = pd.read_csv('exoplanetai_data_column_discription.csv', index_col='S.No')
df_col
```

### Filtering by best default Flag

```python
df = df.loc[df['default_flag'] == 1]
df
```

### Filtering by soltype (only confirmed planet whose data is verified to increase accuracy)

```python
df = df.loc[df['soltype'] == 'Published Confirmed']
df
```

### Filtering by Planet controvery Flag (planet whose existence is under dispute)

```python
df = df.loc[df['pl_controv_flag'] ==0]
df
```

```python
df_p = df_col.set_index('Purpose')
cols_to_drop = df_p.loc['Drop', 'Column Previous Name'].tolist()
df = df.drop(columns=cols_to_drop)
df
```

```python
df_meta_cols = df_p.loc['Metadata','Column Previous Name'].tolist()
df_metadata = df[df_meta_cols]
df_metadata
```

```python
df_feature_cols = df_p.loc['Feature','Column Previous Name'].tolist()
df_feature = df[df_feature_cols]
df_feature.shape
```
