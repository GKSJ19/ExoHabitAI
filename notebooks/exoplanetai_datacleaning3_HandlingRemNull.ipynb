```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('exoplanetai_datacleaning1_finalshape.csv')
df
```

```python
df.isnull().sum()[0:10]
```

```python
df.fillna({
    'hd_name' : 'Na',
    'hip_name' : 'Na',
    'tic_id' : 'Na',
    'gaia_dr2_id' : 'Na',
    'gaia_dr3_id' : 'Na',

},inplace=True)
```

```python
print(df.loc[df['pl_orbper'].isnull(),['pl_orbper','pl_orbsmax','st_mass']].shape)
print(df.loc[df['pl_orbper'].isnull() & df['st_mass'].isnull(),['pl_orbper','pl_orbsmax','st_mass']].shape)
print(df.loc[df['pl_orbper'].isnull() & df['pl_orbsmax'].isnull(),['pl_orbper','pl_orbsmax','st_mass']].shape)
print(df.loc[df['pl_orbper'].isnull() & df['st_mass'].isnull() & df['pl_orbsmax'].isnull(),['pl_orbper','pl_orbsmax','st_mass']].shape)

def orbital_period(row):
    if pd.isnull(row['pl_orbper']):
        if pd.notnull(row['pl_orbsmax']) and pd.notnull(row['st_mass']):
            if row['st_mass'] > 0:
                years = np.sqrt((row['pl_orbsmax']**3) / row['st_mass'])
                return years * 365.25
    return row['pl_orbper']

df['pl_orbper'] = df.apply(orbital_period, axis=1)

# df.dropna(subset=['pl_orbper'], inplace=True)

df.loc[df['pl_orbper'].isnull()].count
```

```python
df['pl_orbpererr1'].fillna(df['pl_orbpererr1'].median, inplace=True)
df['pl_orbpererr2'].fillna(df['pl_orbpererr2'].median, inplace=True)
df['pl_orbperlim'] = df['pl_orbperlim'].fillna(0)
df.loc[df['pl_orbpererr2'].isnull() | df['pl_orbpererr1'].isnull(), ['pl_orbpererr2','pl_orbpererr1']].shape
df.loc[df['pl_orbperlim'].isnull()].shape
```

```python


def kepler_orbsmax(row):
    if pd.isnull(row['pl_orbsmax']):
        if pd.notnull(row['pl_orbper']) and pd.notnull(row['st_mass']):
            if row['st_mass'] > 0:
                years = row['pl_orbper'] / 365.25
                res = (years**2 * row['st_mass'])**(1/3)
                return res
    return row['pl_orbsmax']

df['pl_orbsmax'] = df.apply(kepler_orbsmax, axis=1)
```

```python
df.loc[df['pl_orbsmax'].isnull(), ['st_mass','pl_orbsmax','pl_orbper']].shape
```

```python
df['pl_orbsmaxerr1'].fillna(df['pl_orbsmaxerr1'].median, inplace=True)
df['pl_orbsmaxerr2'].fillna(df['pl_orbsmaxerr2'].median, inplace=True)
df['pl_orbsmaxlim'] = df['pl_orbsmaxlim'].fillna(0)
df.loc[df['pl_orbsmaxerr2'].isnull() | df['pl_orbsmaxerr1'].isnull(), ['pl_orbsmaxerr2','pl_orbsmaxerr1']].shape
df.loc[df['pl_orbsmaxlim'].isnull()].shape
```

```python
df.isnull().sum()[40:45]
```

```python
def handle_pl_rad(row):
    if pd.isnull(row['pl_rade']):
        m = row['pl_masse']
        if pd.notnull(m) and m > 0:
            # Zone 1: Likely Rocky (Mass < 2.0 Earth Masses)
            if m <= 2.0:
                return m**(1/3.7) # Inverse of M = R^3.7
            # Zone 2: Gaseous/Giants
            else:
                return m**(1/2.0) # Inverse of M = R^2.0
                
    return row['pl_rade']

# Apply the imputation
df['pl_rade'] = df.apply(handle_pl_rad, axis=1)
```

```python
df['pl_radeerr1'] = df['pl_radeerr1'].fillna(df['pl_radeerr1'].median())
df['pl_radeerr2'] = df['pl_radeerr2'].fillna(df['pl_radeerr2'].median())

df['pl_radelim'] = df['pl_radelim'].fillna(0)
```
