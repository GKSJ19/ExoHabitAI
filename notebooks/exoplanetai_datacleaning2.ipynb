```python
import pandas as pd
import numpy as np
```

```python
# This creates the variable 'df_raw'
df = pd.read_csv("datacleaning1_final1.csv", low_memory=False)

print(df.shape)
df.head()
```

```python
df.columns = df.columns.str.strip()
```

```python
df.rename(columns={
    "pl_rade": "pl_rad",
    "pl_bmasse": "pl_mass",
    "pl_dens": "pl_dens",
    "pl_eqt": "pl_eqt",
    "pl_orbper": "pl_orbper",
    "pl_orbsmax": "pl_sma",
    "pl_orbeccen": "pl_ecc",
    "pl_insol" : "pl_insol",
    "st_teff": "st_temp",
    "st_lum": "st_lum",
    "st_met": "st_met",
    "st_dens" : "st_dens",
    "st_rad" : "st_rad",
    "st_age" : "st_age",
    "st_mass" : "st_mass",
    "st_spectype": "st_type"
}, inplace=True)
```

```python
features = [
    "pl_rad",
    "pl_mass",
    "pl_dens",
    "pl_eqt",
    "pl_insol",
    "pl_orbper",
    "pl_sma",
    "pl_ecc",
    "st_temp",
    "st_lum",
    "st_met",
    "st_dens",
    "st_rad",
    "st_age",
    "st_mass",
    "st_type"
]

df = df[features].copy()
```

```python
print(df.columns.tolist())
```

```python
df.info()
df.describe()
```

```python
null_report = pd.DataFrame({
    "null_count": df.isna().sum(),
    "null_percent": (df.isna().mean() * 100).round(2)
}).sort_values("null_percent", ascending=False)

null_report
```

```python
df = df.drop(columns=null_report[null_report["null_percent"] > 60].index)
```

```python
for col in df.columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    df[col] = df[col].clip(lower, upper)
```

```python
df.isna().sum()
df.describe()
df.shape
```

```python
total_missing_values = df.isnull().sum().sum()
print("Total missing values:", total_missing_values)
```
