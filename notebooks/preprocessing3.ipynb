```python
def handle_pl_rad(row):
    if pd.isnull(row['pl_rade']):
        m = row['pl_masse']
        if pd.notnull(m) and m > 0:
            # Zone 1: Likely Rocky (Mass < 2.0 Earth Masses)
            if m <= 2.0:
                return m**(1/3.7) # Inverse of M = R^3.7
            # Zone 2: Gaseous/Giants
            else:
                return m**(1/2.0) # Inverse of M = R^2.0
                
    return row['pl_rade']

# Apply the imputation
df['pl_rade'] = df.apply(handle_pl_rad, axis=1)
```

```python
def impute_refined_mass(row):
    # Only act if mass is still NaN after our "harvest"
    if pd.isnull(row['pl_masse']):
        
        # 1. Try Best Mass Proxy
        if pd.notnull(row['pl_bmasse']):
            return row['pl_bmasse']
        
        # 2. Use Radius (Ingredient) to find Mass (Target)
        r = row['pl_rade']
        if pd.notnull(r) and r > 0:
            # Rocky Zone: Mass = Radius^3.7
            if r <= 1.5: 
                return r**3.7
            # Gas Zone: Mass = Radius^2.0
            else:
                return r**2.0
                
    return row['pl_masse']

# Apply the logic
df['pl_masse'] = df.apply(impute_refined_mass, axis=1)
```

```python
def impute_mass_proxies_advanced(row):
    # --- 1. Handling pl_msinie (Minimum Mass) ---
    if pd.isnull(row['pl_msinie']):
        # Step A: Physics Calculation (M * sin(i))
        if pd.notnull(row['pl_masse']) and pd.notnull(row['pl_orbincl']):
            # Convert degrees to radians for np.sin
            incl_rad = np.radians(row['pl_orbincl'])
            row['pl_msinie'] = row['pl_masse'] * np.sin(incl_rad)
        
        # Step B: Direct Proxy (True Mass)
        elif pd.notnull(row['pl_masse']):
            row['pl_msinie'] = row['pl_masse']

    # --- 2. Handling pl_cmasse (Calculated Mass) ---
    if pd.isnull(row['pl_cmasse']):
        # Step A: Direct Proxy (True Mass)
        if pd.notnull(row['pl_masse']):
            row['pl_cmasse'] = row['pl_masse']
            
    return row

# Apply the row-by-row physics logic
df = df.apply(impute_mass_proxies_advanced, axis=1)

# --- 3. Statistical Fallback (Median) ---
# This ensures zero nulls for the ML model without affecting the already-filled values
# df['pl_msinie'] = df['pl_msinie'].fillna(df['pl_msinie'].median())
# df['pl_cmasse'] = df['pl_cmasse'].fillna(df['pl_cmasse'].median())
```

```python
def impute_stellar_radius(row):
    if pd.isnull(row['st_rad']):
        # Tier 1: Inverse Stefan-Boltzmann Calculation
        if pd.notnull(row['st_lum']) and pd.notnull(row['st_teff']):
            # R = sqrt( L / (T/T_sun)^4 )
            temp_ratio_4 = (row['st_teff'] / 5778)**4
            calc_rad = np.sqrt(row['st_lum'] / temp_ratio_4)
            return calc_rad
    return row['st_rad']

# Apply Physics-based Imputation
df['st_rad'] = df.apply(impute_stellar_radius, axis=1)

# Tier 2: Final Median Fallback
# df['st_rad'] = df['st_rad'].fillna(df['st_rad'].median())
# Error handling
df['st_raderr1'] = df['st_raderr1'].fillna(df['st_raderr1'].median())
df['st_raderr2'] = df['st_raderr2'].fillna(df['st_raderr2'].median())

# Limit Flag (0 = Exact measurement)
df['st_radlim'] = df['st_radlim'].fillna(0)
```

```python
def calculate_stellar_luminosity(row):
    if pd.isnull(row['st_lum']):
        # Tier 1: Physics Calculation
        if pd.notnull(row['st_rad']) and pd.notnull(row['st_teff']):
            # Normalized Stefan-Boltzmann formula
            # L = R^2 * (T/T_sun)^4
            calc_lum = (row['st_rad']**2) * ((row['st_teff'] / 5778)**4)
            return calc_lum
    return row['st_lum']

# Apply the physics-first calculation
df['st_lum'] = df.apply(calculate_stellar_luminosity, axis=1)

# Tier 2: Statistical Fallback (Median)
# Only fills if the star is missing BOTH radius and temperature
# df['st_lum'] = df['st_lum'].fillna(df['st_lum'].median())
# Handle Uncertainty using Median
df['st_lumerr1'] = df['st_lumerr1'].fillna(df['st_lumerr1'].median())
df['st_lumerr2'] = df['st_lumerr2'].fillna(df['st_lumerr2'].median())

# Handle Limit Flag
df['st_lumlim'] = df['st_lumlim'].fillna(0)
```

```python
def calculate_insolation(row):
    if pd.isnull(row['pl_insol']):
        # Tier 1: Physics Calculation
        if pd.notnull(row['st_lum']) and pd.notnull(row['pl_orbsmax']):
            if row['pl_orbsmax'] > 0:
                # Flux follows the Inverse Square Law
                calc_insol = row['st_lum'] / (row['pl_orbsmax']**2)
                return calc_insol
    return row['pl_insol']

df['pl_insol'] = df.apply(calculate_insolation, axis=1)

# Handle Limit Flag
df['pl_insollim'] = df['pl_insollim'].fillna(0)
```

```python
def calculate_equilibrium_temp(row):
    if pd.isnull(row['pl_eqt']):
        # Tier 1: Physics Calculation
        # We need st_teff, st_rad (in solar units), and pl_orbsmax (in AU)
        if pd.notnull(row['st_teff']) and pd.notnull(row['st_rad']) and pd.notnull(row['pl_orbsmax']):
            if row['pl_orbsmax'] > 0:
                # Standard albedo assumption (0.3)
                albedo = 0.3
                # R_star needs to be converted to AU for the formula, 
                # but we can use the ratio-based version:
                # T_eq = T_eff * sqrt(R_star / (2 * semi_major_axis_in_Solar_Radii))
                # 1 AU = 215 Solar Radii
                a_solar_radii = row['pl_orbsmax'] * 215
                calc_temp = row['st_teff'] * ((row['st_rad'] / (2 * a_solar_radii))**0.5) * ((1 - albedo)**0.25)
                return calc_temp
    return row['pl_eqt']

df['pl_eqt'] = df.apply(calculate_equilibrium_temp, axis=1)

# Tier 2: Statistical Fallback (Median)
# df['pl_eqt'] = df['pl_eqt'].fillna(df['pl_eqt'].median())
```

```python
def calculate_inclination(row):
    if pd.isnull(row['pl_orbincl']):
        # Tier 1: Derive from Impact Parameter (b) if available
        # b = (a * cos(i)) / R_star
        if pd.notnull(row['pl_imppar']) and pd.notnull(row['pl_orbsmax']) and pd.notnull(row['st_rad']):
            # Convert a (AU) to Solar Radii (215)
            a_rs = row['pl_orbsmax'] * 215
            cos_i = (row['pl_imppar'] * row['st_rad']) / a_rs
            # Clamp cos_i between -1 and 1 to avoid math errors
            cos_i = max(min(cos_i, 1), -1)
            return np.degrees(np.arccos(cos_i))
            
    return row['pl_orbincl']

df['pl_orbincl'] = df.apply(calculate_inclination, axis=1)

# Tier 2: Median Fallback
# df['pl_orbincl'] = df['pl_orbincl'].fillna(df['pl_orbincl'].median())
```

```python
def calculate_impact_parameter(row):
    if pd.isnull(row['pl_imppar']):
        # Tier 1: Physics Calculation
        # b = (a * cos(i)) / R_star
        if pd.notnull(row['pl_orbincl']) and pd.notnull(row['pl_orbsmax']) and pd.notnull(row['st_rad']):
            # Convert a (AU) to Solar Radii (215) to match st_rad units
            a_rs = row['pl_orbsmax'] * 215
            incl_rad = np.radians(row['pl_orbincl'])
            
            calc_b = (a_rs * np.cos(incl_rad)) / row['st_rad']
            return calc_b
    return row['pl_imppar']

df['pl_imppar'] = df.apply(calculate_impact_parameter, axis=1)

# Tier 2: Median Fallback
df['pl_imppar'] = df['pl_imppar'].fillna(df['pl_imppar'].median())
```

```python
def calculate_transit_depth(row):
    if pd.isnull(row['pl_trandep']):
        # Tier 1: Physics Calculation
        if pd.notnull(row['pl_rade']) and pd.notnull(row['st_rad']):
            # Convert st_rad to Earth units (109.1)
            r_star_earth = row['st_rad'] * 109.1
            # Ratio of areas squared (converted to percentage)
            calc_depth = (row['pl_rade'] / r_star_earth)**2 * 100
            return calc_depth
    return row['pl_trandep']

df['pl_trandep'] = df.apply(calculate_transit_depth, axis=1)

# Tier 2: Median Fallback
# df['pl_trandep'] = df['pl_trandep'].fillna(df['pl_trandep'].median())
```

```python
def calculate_transit_duration(row):
    if pd.isnull(row['pl_trandur']):
        # Tier 1: Geometric Physics Approximation
        if pd.notnull(row['pl_orbper']) and pd.notnull(row['st_rad']) and \
           pd.notnull(row['pl_orbsmax']) and pd.notnull(row['pl_imppar']):
            
            # Convert a (AU) to Solar Radii (215)
            a_rs = row['pl_orbsmax'] * 215
            # P is in days, st_rad is in Solar units, a is in Solar units
            # Formula output in days, so multiply by 24 for hours
            # We use the factor (24 / pi)
            term = (row['st_rad'] / a_rs) * np.sqrt(max(0, 1 - row['pl_imppar']**2))
            calc_dur = (row['pl_orbper'] * 24 / np.pi) * term
            return calc_dur
    return row['pl_trandur']

df['pl_trandur'] = df.apply(calculate_transit_duration, axis=1)

# Tier 2: Median Fallback
# df['pl_trandur'] = df['pl_trandur'].fillna(df['pl_trandur'].median())
```

```python
def calculate_rv_amplitude(row):
    if pd.isnull(row['pl_rvamp']):
        # Tier 1: Physics Calculation (RV Semi-amplitude K)
        # Using a simplified version for Solar/Earth units:
        # K â‰ˆ (28.4 m/s) * (Mp * sin(i) / P^(1/3)) * (Ms + Mp)^(-2/3)
        if pd.notnull(row['pl_bmasse']) and pd.notnull(row['pl_orbper']) and pd.notnull(row['st_mass']):
            # P in days, Mp in Earth masses, Ms in Solar masses
            # Convert Mp to Jupiter masses (Mp / 317.8) for standard K constant
            mp_jup = row['pl_bmasse'] / 317.8
            p_year = row['pl_orbper'] / 365.25
            
            # Using the standard RV amplitude approximation
            # Note: sin(i) is assumed 1 (edge-on) if pl_orbincl is missing or 90
            sin_i = np.sin(np.radians(row['pl_orbincl'])) if pd.notnull(row['pl_orbincl']) else 1.0
            
            calc_k = (28.4329 / np.sqrt(1 - row['pl_orbeccen']**2)) * \
                     (mp_jup * sin_i) * \
                     (row['st_mass']**(-2/3)) * \
                     (p_year**(-1/3))
            return calc_k
    return row['pl_rvamp']

# Apply Physics-based Imputation
df['pl_rvamp'] = df.apply(calculate_rv_amplitude, axis=1)

# Tier 2: Median Fallback
# df['pl_rvamp'] = df['pl_rvamp'].fillna(df['pl_rvamp'].median())
```

```python
def map_spectral_type(row):
    if pd.isnull(row['st_spectype']):
        temp = row['st_teff']
        if pd.notnull(temp):
            # Standard Harvard Classification Mapping
            if temp >= 30000: return 'O'
            elif temp >= 10000: return 'B'
            elif temp >= 7500: return 'A'
            elif temp >= 6000: return 'F'
            elif temp >= 5200: return 'G'
            elif temp >= 3700: return 'K'
            elif temp >= 2400: return 'M'
            else: return 'L/T (Brown Dwarf)'
    return row['st_spectype']

# Apply the mapping logic
df['st_spectype'] = df.apply(map_spectral_type, axis=1)

# Final Fallback for any remaining nulls
df['st_spectype'] = df['st_spectype'].fillna('Unknown')
```

```python
def calculate_stellar_mass(row):
    if pd.isnull(row['st_mass']):
        # Tier 1: Gravity-Radius Calculation
        if pd.notnull(row['st_logg']) and pd.notnull(row['st_rad']):
            # Convert logg to absolute gravity
            g_abs = 10**row['st_logg']
            # M = (g * R^2) / g_sun
            calc_mass = (g_abs * (row['st_rad']**2)) / 27422.5
            return calc_mass
    return row['st_mass']

# Apply Physics-based Imputation
df['st_mass'] = df.apply(calculate_stellar_mass, axis=1)

# Tier 2: Statistical Fallback (Median)
# df['st_mass'] = df['st_mass'].fillna(df['st_mass'].median())
```

```python
def calculate_stellar_logg(row):
    if pd.isnull(row['st_logg']):
        # Tier 1: Physics Calculation
        if pd.notnull(row['st_mass']) and pd.notnull(row['st_rad']):
            if row['st_rad'] > 0:
                # Calculate absolute gravity relative to Sun
                g_val = (row['st_mass'] / (row['st_rad']**2)) * 27422.5
                return np.log10(g_val)
    return row['st_logg']

df['st_logg'] = df.apply(calculate_stellar_logg, axis=1)

# Tier 2: Median Fallback
# df['st_logg'] = df['st_logg'].fillna(df['st_logg'].median())
```

```python
def calculate_stellar_density(row):
    if pd.isnull(row['st_dens']):
        # Tier 1: Physics Calculation
        if pd.notnull(row['st_mass']) and pd.notnull(row['st_rad']):
            if row['st_rad'] > 0:
                # Density in g/cm3
                return 1.41 * (row['st_mass'] / (row['st_rad']**3))
    return row['st_dens']

df['st_dens'] = df.apply(calculate_stellar_density, axis=1)

# Tier 2: Median Fallback
# df['st_dens'] = df['st_dens'].fillna(df['st_dens'].median())
```

```python
# 1. First, fill the base components with Median to provide the ingredients for the sum
df['sy_pmra'] = df['sy_pmra'].fillna(df['sy_pmra'].median())
df['sy_pmdec'] = df['sy_pmdec'].fillna(df['sy_pmdec'].median())

# 2. Tier 1: Physics-based calculation for Total Proper Motion
def calculate_total_pm(row):
    if pd.isnull(row['sy_pm']):
        # Vector Sum: sqrt(pmra^2 + pmdec^2)
        if pd.notnull(row['sy_pmra']) and pd.notnull(row['sy_pmdec']):
            return np.sqrt(row['sy_pmra']**2 + row['sy_pmdec']**2)
    return row['sy_pm']

df['sy_pm'] = df.apply(calculate_total_pm, axis=1)

# Tier 2: Statistical Fallback for sy_pm
# df['sy_pm'] = df['sy_pm'].fillna(df['sy_pm'].median())
```

```python
def reconcile_distance_parallax(row):
    # 1. Fill Distance from Parallax
    if pd.isnull(row['sy_dist']):
        if pd.notnull(row['sy_plx']) and row['sy_plx'] > 0:
            return 1000 / row['sy_plx']
            
    # 2. Fill Parallax from Distance
    if pd.isnull(row['sy_plx']):
        if pd.notnull(row['sy_dist']) and row['sy_dist'] > 0:
            row['sy_plx'] = 1000 / row['sy_dist']
            
    return row['sy_dist']

# Apply the cross-reconciliation
df['sy_dist'] = df.apply(reconcile_distance_parallax, axis=1)

# Second pass for the Parallax column specifically
def fill_plx(row):
    if pd.isnull(row['sy_plx']):
        if pd.notnull(row['sy_dist']) and row['sy_dist'] > 0:
            return 1000 / row['sy_dist']
    return row['sy_plx']

df['sy_plx'] = df.apply(fill_plx, axis=1)

# Tier 2: Final Median Fallback for both
# df['sy_dist'] = df['sy_dist'].fillna(df['sy_dist'].median())
# df['sy_plx'] = df['sy_plx'].fillna(df['sy_plx'].median())
```
