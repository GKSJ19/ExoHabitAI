### Drop unused columns
Remove columns flagged as `Drop` in the description metadata.

```python
def drop_columns_by_purpose(df, df_col, drop_label='Drop'):
    df_col = df_col.set_index('Purpose')

    cols = (
        df_col
        .loc[drop_label, 'Column Previous Name']
        .dropna()
        .unique()
        .tolist()
    )

    cols = [c for c in cols if c in df.columns]

    return df.drop(columns=cols), cols


df, dropped = drop_columns_by_purpose(df, df_col)
print("Dropped:", dropped)
print("Final shape:", df.shape)
```

```python
df_p = df_col.set_index('Purpose')
```

### Split metadata columns
Extract metadata-only columns for reference and inspection.

```python
def extract_columns_by_purpose(df, df_p, purpose):
    if purpose not in df_p.index:
        raise ValueError(f"{purpose} not found in Purpose index")

    cols = (
        df_p.loc[purpose, 'Column Previous Name']
        .dropna()
        .unique()
        .tolist()
    )

    cols = [c for c in cols if c in df.columns]
    return df.loc[:, cols], cols


df_metadata, meta_cols = extract_columns_by_purpose(df, df_p, 'Metadata')
print("Metadata shape:", df_metadata.shape)
```

### Split feature columns
Isolate feature columns for model training inputs.
